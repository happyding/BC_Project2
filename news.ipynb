```python
#get news data with the keyword "oil" from 2020-12-17 to 2020-12-31.
oil2_data = newsapi.get_everything(q='oil',
                                      sources='business insider, reuters, cnn, the-wall-street-journal, financial-post, bloomberg',
                                      domains='businessinsider.com,reuters.com, us.cnn.com, wsj.com, business.financialpost.com, bloomberg.com',
                                      from_param = "2020-12-17T00:00:00",
                                      to = "2020-12-31T23:59:59",
                                      language='en',
                                      sort_by='relevancy',
                                      page_size=100,
                                      
                                )
oil2_data['totalResults']
```

```python
#create the dataframe with the articles from oil_data
oil2_df = pd.DataFrame(oil2_data['articles'])

#convert object column to datetime
oil2_df['publishedAt'] = pd.to_datetime(oil2_df['publishedAt'], infer_datetime_format=True)

#drop the three unnecessary columns
oil2_df.drop(['urlToImage','description','author'], axis=1, inplace=True)

oil2_df
```

```python
#concatenate the two dataframes
alll = pd.concat([oil2_df,oil_df], axis=0, join='inner')

#shorten the text under the "content" column to 500 characters and save into a new column
alll['Content'] = alll['content'].str[:500]

#create a new column "HourID" by simplifying the "publishedAt" column
alll['HourID'] = alll.publishedAt.dt.strftime('%Y%m%d%H')

#grouby the dataframe by "HourID" and "publishedAt"
combined_df = alll.groupby(["HourID","publishedAt"])
combined_df = combined_df.first()
combined_df
```

```python
#sort the column "publishedAt" in ascending order and reset the index
combined_df.sort_values('publishedAt', ascending=True).reset_index()

#cut out news that include the word "palm oil", "uk", and "india"
combined_df = combined_df[~combined_df.title.str.contains("palm oil")]
combined_df = combined_df[~combined_df.url.str.contains('uk','india')]

#drop duplicated news
combined_df.drop_duplicates(subset='title',keep='first', inplace=True)
combined_df
```
