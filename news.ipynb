```python
from newsapi import NewsApiClient
import datetime as dt
import pandas as pd
from datetime import datetime
```

```python
newsapi = NewsApiClient(api_key='0061e663f067480d88f5d0b881de99bb')
```

```python
usd_data = newsapi.get_everything(q='USD',
                                      sources='business insider, reuters',
                                      domains='businessinsider.com,reuters.com',
                                      from_param='2020-12-11',
                                      to='2021-01-11',
                                      language='en',
                                      sort_by='relevancy',
                                      page=2)
```

```python
usd_articles = usd_data['articles']
for x,y in enumerate(usd_articles):
        print(f'{x}   {y["title"]}')
```

```python
usd_df = pd.DataFrame(usd_articles)

usd_df['publishedAt'] = pd.to_datetime(usd_df['publishedAt'], infer_datetime_format=True)
usd_df = usd_df.set_index('publishedAt')
usd_df.sort_index(ascending=True, inplace=True)
usd_df.head()
```
